```python
import numpy as np
from matplotlib import pyplot as plt
from paths import *
from photometry import *

setup_logging()

region = 'N'
path_data = path_data(region)
path_randoms = path_randoms(region)
path_mctool = path_mctool(region)
```

```python
data = TargetSelection.load_objects(path_data,tracer='ELG',region=region)
data = data.apply_maskbit()
randoms = TargetSelection.load_objects(path_randoms,tracer='ELG',region=region)
randoms = randoms.apply_maskbit()
```

```python
dens = BrickDensity(ref=randoms)
dens.set_randoms(randoms=randoms)
dens.set_properties()
dens.plot_property_map(prop='EBV',title='EBV')
dens.set_data(data=data)
dens.plot_density_map(title='target density',vmin=0,vmax=2)
dens.plot_property_map(prop='GALDEPTH_G',title='g-depth')
```

```python
map = Catalogue.load(path_hp)
dens = HealpixDensity(map=map)
dens.set_randoms(randoms=randoms)
dens.set_properties()
dens.set_data(data=data)
dens.plot_density_map(title='target density',vmin=0,vmax=2)

dens.plot_density_variations(prop='EBV',histos=[dens],xedges={'range':[0,0.2]})
dens.plot_density_variations(prop='STARDENS',histos=[dens],xedges={'range':[None,4000]})
for field in ['GALDEPTH','PSFSIZE']:
    fig,lax = plt.subplots(ncols=3,nrows=1,sharex=False,sharey=True,figsize=(12,4))
    lax = lax.flatten()
    for iax,prop in enumerate(['{}_{}'.format(field,b) for b in ['G','R','Z']]):
        dens.plot_density_variations(ax=lax[iax],prop=prop,histos=[dens],xedges={'quantiles':[0.01,0.99]})
    plt.show()
densebv = dens[dens.properties['EBV']<0.04]
dens.plot_density_variations(prop='SN2TOT_FLAT',others=[densebv],histos=[dens,densebv],var_kwargs={'labels':[None,'EBV < 0.04']},xedges={'range':[100.,1500.]})
dens.plot_density_variations(prop='SN2TOT_G',others=[densebv],histos=[dens,densebv],var_kwargs={'labels':[None,'EBV < 0.04']},xedges={'range':[50.,1000.]})
dens.properties['fracPSF'] = dens.properties.get_statistics(statistic='frac',ids_catalogue=dens.get_bricks(data),values_catalogue=data.mask_morphtype('PSF'))
densebv.properties['fracPSF'] = densebv.properties.get_statistics(statistic='frac',ids_catalogue=densebv.get_bricks(data),values_catalogue=data.mask_morphtype('PSF'))
dens.plot_density_variations(prop='fracPSF',others=[densebv],histos=[dens,densebv],var_kwargs={'labels':[None,'EBV < 0.04']},xedges={'range':[0.,0.4]})
```

```python
map = Catalogue.load(path_mctool)
dens = HealpixDensity(map=map)
dens.set_randoms(randoms=randoms)
dens.set_properties()
dens.set_data(data=data)
weights = dens.properties['MCEFF'][:]
weights[weights==0.] = 1. # to avoid regions of psfdepth == 0
densmc = dens / weights
m = LinearModel(density=dens)
m.fit(props=['EBV'])
m.predict(catalogue=dens.properties,key_efficiency='LINEBV')
denslinebv = dens / dens.properties['LINEBV']
m.fit(props=['STARDENS'])
m.predict(catalogue=dens.properties,key_efficiency='LINSTAR')
denslinstar = dens / dens.properties['LINSTAR']
m.fit(props=['EBV','STARDENS'])
m.predict(catalogue=dens.properties,key_efficiency='LINEBVSTAR')
denslinebvstar = dens / dens.properties['LINEBVSTAR']

dens.plot_density_variations(prop='EBV',others=[densmc,denslinebv,denslinstar],histos=[dens],var_kwargs={'labels':[None,'+ MC weights','+ EBV weights','+ STARDENS weights']},xedges={'range':[0.,.2]})
dens.plot_density_variations(prop='STARDENS',others=[densmc,denslinebv,denslinebvstar],histos=[dens],var_kwargs={'labels':[None,'+ MC weights','+ EBV weights','+ STARDENS, EBV weights']},xedges={'quantiles':[0.01,0.99]})
for field in ['GALDEPTH','PSFSIZE']:
    fig,lax = plt.subplots(ncols=3,nrows=1,sharex=False,sharey=True,figsize=(12,4))
    lax = lax.flatten()
    for iax,prop in enumerate(['{}_{}'.format(field,b) for b in ['G','R','Z']]):
        dens.plot_density_variations(ax=lax[iax],prop=prop,others=[densmc,denslinebv,denslinebvstar],histos=[dens],var_kwargs={'labels':[None,'+ MC weights','+ EBV weights','+ STARDENS, EBV weights']},xedges={'quantiles':[0.01,0.99]})
    plt.show()
```

```python
from scipy import special

fig,ax = plt.subplots(ncols=1,nrows=1,sharex=False,sharey=False,figsize=(6,6))
dens.plot_density_variations(ax=ax,others=[],prop='EBV',histos=[dens],var_kwargs={'labels':'DR9'},leg_kwargs={},xedges={'quantiles':[0.1,0.9]})
ebv = np.linspace(*ax.get_xlim(),num=10)
model = 1. - .25/.15*ebv # Ashley's fit
norm = 1.04 # just for plotting purposes
model = model/np.mean(model)/norm - 1.

truth = TargetSelection.load_objects(path_truth,tracer='ELG',region=None,case_sensitive=False)
truth.set_flux_from_mag(key='FLUX')
truth['REDSHIFT'] = truth['hsc_mizuki_photoz_best']
truth.region = region
mask = np.all([~np.isnan(truth[b]) for b in truth.bands],axis=0)
truth = truth[mask]

correlation = np.ones((3,3))*.5 #just given them all correlation of 0.5 for now
np.fill_diagonal(correlation,1.)
std = np.array([0.03,0.05,0.1])
flux_covariance = correlation*std*std[:,None]
efficiency = []
for ebv_ in ebv:
    mc = MCTool(truth=truth,seed=42)
    mc.set_sel_params(ebvfac=1,Rv=None,sn_band_min=6,sn_flat_min=None,sn_red_min=None)
    mc.set_sim_params(flux_covariance=flux_covariance,flux_adbias=0.,flux_mulbias=1.)
    mc.sim['EBV'] = ebv_
    mc.sim.set_estimated_transmission(key='EMW_TRANSMISSION')
    mc.sim.set_estimated_transmission(key='MW_TRANSMISSION')
    transmission = 10.**(-0.4*0.6*TargetSelection.EXT_COEFFS['G']*mc.sim['EBV'])
    transmission = transmission + (1.-transmission)*special.erf(2.*mc.truth['FLUX_G'])
    mc.sim['MW_TRANSMISSION_G'] *= transmission
    mc()
    efficiency.append(mc.get_efficiency())
efficiency = np.array(efficiency)/np.mean(efficiency)/norm - 1.

ax.plot(ebv,model,label='empirical fit',color='k')
ax.plot(ebv,efficiency,label='MC prediction',color='r')
ax.legend()
plt.show()
```
